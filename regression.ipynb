# Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Advertising.csv")
```

```python
x = data.values[:,2]
y = data.values[:,4]
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y,marker = 'o')
```

```python
def predict(X,beta0, beta1):
    return beta1*X + beta0


def cost_function(X,Y,beta0,beta1):
    m = len(X)
    sum_error = 0
    for i in range(m):
        sum_error += (Y[i] -(beta1*X[i]+beta0))**2
    return sum_error/(2*m)


def update (X,Y,beta0,beta1, learning_rate):
    m = len(X)
    temp0 = 0
    temp1 = 0
    for i in range(m):
        temp0+= (beta1*X[i] + beta0 - Y[i])
        temp1+= (beta1*X[i] + beta0 - Y[i])*X[i]
    beta0 -= (temp0/m) * learning_rate
    beta1 -= (temp1/m) * learning_rate
    return beta0, beta1


def train(X,Y,beta0,beta1, learning_rate, iter):
    cos_his = []
    for i in range(iter):
        beta0, beta1 = update(X,Y,beta0,beta1, learning_rate)
        cost = cost_function(X,Y,beta0,beta1)
        cos_his.append(cost)
    return beta0,beta1,cos_his
```

```python
beta0,beta1,cost = train(x,y,0.03,0.0014,0.001,60)
```

```python
predict(19,beta0,beta1)
```

```python
plt.plot(cost)
```

```python
def plot_regression_line(x, y, beta0, beta1): 
    # plotting the actual points as scatter plot 
    plt.scatter(x, y, color = "m", 
               marker = "o", s = 30) 
  
    # predicted response vector 
    y_pred = beta0 + beta1*x 
  
    # plotting the regression line 
    plt.plot(x, y_pred, color = "g") 
  
    # putting labels 
    plt.xlabel('x') 
    plt.ylabel('y') 
  
    # function to show plot 
    plt.show()
```

```python
plot_regression_line(x,y,beta0,beta1)
```
